```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/DuaLipa.csv')
```

```python
df.head()
```

```python
print(df.columns)
```

```python
all_lyrics = ' '.join(df['Lyric'].dropna().astype(str)).lower()
print(all_lyrics[:500])
```

```python
import string
from collections import Counter
import matplotlib.pyplot as plt
translator = str.maketrans('', '', string.punctuation)
cleaned_lyrics = all_lyrics.translate(translator)

words = cleaned_lyrics.split()


word_counts = Counter(words)


sorted_word_counts = word_counts.most_common()


ranks = [rank + 1 for rank in range(len(sorted_word_counts))]
frequencies = [freq for word, freq in sorted_word_counts]

plt.figure(figsize=(10, 6))
plt.plot(ranks, frequencies, marker='o', linestyle='-', color='b')
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Rank of Word')
plt.ylabel('Frequency of Word')
plt.title('Zipf\'s Law: Word Frequency vs Rank')
plt.grid(True)
plt.show()
```
